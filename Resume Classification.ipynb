```python
import glob

# Set the directory path where the docx files are located
directory_path = "C:/Users/lenovo/Downloads/Resumes/Resumes/SQL Developer Lightning insight"

# Define the pattern for finding docx files
pattern = directory_path + "/*.docx"

# Use the glob function to find all the docx files
docx_files = glob.glob(pattern)

# Print the list of docx files found
print(docx_files)
```

```python
#docx_files = glob.glob(pattern)
import docx
# Loop over each file and perform some operation on it
for docx_file in docx_files:
    # Open the document file
    doc = docx.Document(docx_file)
    
    # Extract the text from the document
    text = '\n'.join([paragraph.text for paragraph in doc.paragraphs])
    
    # Print the extracted text
    print(text)
```
